## Now on Second Data 

```python
import pandas as pd
import numpy as np  
import seaborn as sns
```

Loading dataset...

```python
df=sns.load_dataset('planets')
df.head()
```

Cheaking Missing Values

```python
df.isnull().sum().sort_values(ascending=False)
```

Mass has high missing values we drop it

```python
df.drop('mass',axis=1,inplace=True)
df.isnull().sum().sort_values(ascending=False)
```

For distance we take Median

```python
mean_distance=df['distance'].mean()
df['distance'].fillna(mean_distance,inplace=True)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

orbital period is numeric we take median 

```python
orbital_period_median=df['orbital_period'].median()
df['orbital_period'].fillna(orbital_period_median,inplace=True)
```

```python
df.isnull().sum().sort_values(ascending=False)
```
