Finding and removing missing Values using K nearest neighbour

we convert values into numerical firstly

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Firstly laod dataset

```python
df=sns.load_dataset('titanic')
df.head()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

Now convert all categorical values into numerical

```python
from sklearn.preprocessing import LabelEncoder

columns=['sex','embarked','class','who','adult_male','deck','embark_town','alive']
label_encoders={} #dictionary to hold label encoders 4 each column

for col in columns:
    le=LabelEncoder()
    df[col]=le.fit_transform(df[col])
    label_encoders[col]=le
df.head()
```

values are converted to numerical form

Now Cheaking Missing Values

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
# using KNN Imputer to fill missing values
from sklearn.impute import KNNImputer
imputer=KNNImputer(n_neighbors=5)

df['age']=imputer.fit_transform(df[['age']])

df.isnull().sum().sort_values(ascending=False)
```

data is cleaned using K nearest Neighbour Algorithm

now convert data set back in original form 

```python
for col in columns:
    le=label_encoders[col]
    df[col]=le.inverse_transform(df[col])
df.head()
```

data is converted back in orignal form
