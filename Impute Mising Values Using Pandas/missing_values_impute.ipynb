**Imputing Missing Values Using Pandas**

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

Loading Dataset...

```python
df=sns.load_dataset('titanic')
df.head()
```

Cheaking Missing Values

```python
df.isnull().sum().sort_values(ascending=False)
```

droping Deck column

```python
df.drop('deck',axis=1 ,inplace=True)
df.isnull().sum().sort_values(ascending=False)
```

For Age column we take median 

```python
median_age=df['age']=df['age'].median()
df['age'].fillna(median_age,inplace=True)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

for embarked,embarked_town  we tke mode

```python
embareked_mode=df['embarked'].mode()[0]
df['embarked'].fillna(embareked_mode,inplace=True)
```

```python
embarked_town_mode=df['embark_town'].mode()[0]
df['embark_town'].fillna(embarked_town_mode,inplace=True)
```

```python
df.isnull().sum().sort_values(ascending=False)
```
