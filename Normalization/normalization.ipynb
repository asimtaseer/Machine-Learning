# Normalization

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

Data set

```python
np.random.seed(0)
data=np.random.exponential(scale=2, size=1000)
df=pd.DataFrame(data, columns=['value'])
df.head()
```

Normalization

```python
from sklearn.preprocessing import power_transform
powered = power_transform(df, method='box-cox')
df['powered_value'] = powered
df.head()
```

```python
y=power_transform(df[['value']],method='yeo-johnson')
df['y_johnson']=y
df.head()
```

```python
sns.histplot(df['value'], kde=True)
```

```python
for col in df.columns:
    sns.histplot(df[col], kde=True)
    plt.show()
```
