# Min Max scaler

```python
import pandas as pd
from sklearn.preprocessing import  MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import maxabs_scale
```

loading dataset....

```python
data={
    'age': [25, 45, 35, 50, 23],
    'salary': [50000, 100000, 75000, 120000, 48000],
    'score': [200, 400, 300, 500, 250]
}

df = pd.DataFrame(data)
df.head()
```

Min max, max abs, standard scaler

```python
standard_scaler = StandardScaler()
min_max_scaler = MinMaxScaler()
max_abs_scaler = maxabs_scale\

df_standard_scaled = standard_scaler.fit_transform(df)
df_min_max_scaled = min_max_scaler.fit_transform(df)
df_max_abs_scaled = max_abs_scaler(df)

df_standard_scaled = pd.DataFrame(df_standard_scaled, columns=df.columns)
df_min_max_scaled = pd.DataFrame(df_min_max_scaled, columns=df.columns) 
df_max_abs_scaled = pd.DataFrame(df_max_abs_scaled, columns=df.columns)
print("Standard Scaled Data:\n", df_standard_scaled)
print("\nMin-Max Scaled Data:\n", df_min_max_scaled)
print("\nMax-Abs Scaled Data:\n", df_max_abs_scaled)
```
