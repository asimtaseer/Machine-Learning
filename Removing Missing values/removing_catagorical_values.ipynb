Removing all the catagorical values in the dataset


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

loading data set

```python
df=sns.load_dataset('titanic')
df.head()
```

```python

columns=['sex','embarked','class','who','adult_male','deck','embark_town','alive']
label_encoders={} #dictionary to hold label encoders 4 each column
for col in columns:
    le=LabelEncoder()
    df[col]=le.fit_transform(df[col])
    label_encoders[col]=le
df.head()
```

There is no catagorical value in the data set
